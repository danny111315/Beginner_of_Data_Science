#### **流程控制**



```python
age = 18

if age > 18:    #58. if
  print("你成年了")


age = 16

if age >= 18:   #58. if......else
  print("你已經成年了")
else:
  print("你未成年")


score  = 85    #59. if......else
if score >= 90:
  print("優秀")
elif score >= 80:
  print("良好")
else:
  print("加油")

age = 18      #60. 一次表達式
result = "成年" if age >= 18 else "未成年"
print(result)
```

#### **迴圈**

```python
count = 0         #63. 【while】 迴圈 ，遍歷一個範圍內的數字

while count < 5:
  print("count =", count)
  count += 1

for i in range(3):    #64. 【for】 迴圈 ，遍歷一個範圍內的數字
  print("你好")

for i in range(2,11,2):
  print(i)         #65. 【for】 迴圈 ，用range()指定範圍、間隔


fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
  print(fruit)

for i in range(10):    #66. 【break】語法 ，斷掉迴圈執行
  if i == 5:
    break
  print("i =", i)

for i in range(10):
  if i % 2 == 0:    #67. 【continue】語法 ，掠過指定條件
    continue
  print("odd number =", i)

square = [x * x for x in range(5)]
print(square)         #68. 推導表達式 ，允許在一行生成列表、字典、集合


odd_squares = [x * x for x in range(10) if x % 2 != 0]
print(odd_squares)      #69. 推導表達式 配合if 語法 ，推成列表

word_lengths = {word: len(word) for word in["apple", "banana", "cherry"]}
print(word_lengths)      #69. 推導表達式 配合if 語法 ，推成字典格式
```

#### **函式**


```python
def add(a,b):    #70. 推導表達式 配合if 語法 ，推成字典格式
  return a + b
result = add(3,5)
print(result)


def func():
  local_var = "我是區域變數" #變數介紹
  print(local_var)

global_var = "我是全域變數"
func()
print(global_var)
```

```python
x = 5 # 全域

def multiply_by_two():
  x = 2   #局域
  print(x * 2)   #局域

multiply_by_two()
print(x)   #全域
```

```python
square = lambda x: x * x   #71. lambda (匿名函數)
print(square(5))

numbers = [1,2,3,4,5]
squares = list(map(lambda x: x * x, numbers))
print(squares)
```
